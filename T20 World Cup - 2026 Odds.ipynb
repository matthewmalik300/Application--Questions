```python
import numpy as np
import pandas as pd

bookmakers = ['Bet365', 'William Hill', 'Betfred', 'Ladbrokes', 'Paddy Power']

india =       [11/8,  11/8, 13/8, 5/4, 6/4]
australia =   [4/1,   4/1,  5/1,  4/1, 4/1]
england =     [9/2,   9/2,  11/2, 5/1, 11/2]
south_africa =[11/2,  11/2, 11/2, 6/1, 9/2]
new_zealand = [16/1,  14/1, 12/1, 14/1, 16/1]

odds_array = np.array([
    india,
    australia,
    england,
    south_africa,
    new_zealand
], dtype=float)

teams = ['India', 'Australia', 'England', 'South Africa', 'New Zealand']

df = pd.DataFrame(odds_array, index=teams, columns=bookmakers)

```

```python
Decimal_odds = df + 1
```

```python
Decimal_odds["Mean"] = Decimal_odds.mean(axis=1)
```
