Poisson CDF Function Generation

```python
import math

def pois_pmf(k, lam):
    return math.exp(-lam) * lam**k / math.factorial(k)

def match_probs(lam_M, lam_L, K=10):
    # pmfs
    pM = [pois_pmf(k, lam_M) for k in range(K+1)]
    pL = [pois_pmf(k, lam_L) for k in range(K+1)]

    # CDFs
    FM = []
    FL = []
    cum_M = cum_L = 0.0
    for k in range(K+1):
        cum_M += pM[k]
        cum_L += pL[k]
        FM.append(cum_M)
        FL.append(cum_L)

    # P(draw)
    p_draw = sum(pM[k] * pL[k] for k in range(K+1))

    # P(City > Liverpool)
    p_M_win = sum(pM[m] * FL[m-1] for m in range(1, K+1))

    # P(Liverpool > City)
    p_L_win = sum(pL[l] * FM[l-1] for l in range(1, K+1))

    return p_M_win, p_draw, p_L_win
```

```python
def recency_weighted_xg(xg_old, n_old, xg_new, n_new, w):
    eff_n_new = w * n_new
    return (n_old * xg_old + eff_n_new * xg_new) / (n_old + eff_n_new)
```

```python
import pandas as pd
import numpy as np

w_values = np.linspace(1, 5, 9)

def recency_weighted_xg(xg_old, n_old, xg_new, n_new, w):
    eff_n_new = w * n_new
    return (n_old * xg_old + eff_n_new * xg_new) / (n_old + eff_n_new)

Liv_2425 = df_xG_xGA_2425["xG"][0]
City_2425 = df_xG_xGA_2425["xG"][1]
n_2425 = 38

Liv_2526 = df_xG_xGA_2526["xG"][0]
City_2526 = df_xG_xGA_2526["xG"][1]
n_2526 = 10

rows = []
for w in w_values:
    lambda_city = recency_weighted_xg(City_2425, n_2425, City_2526, n_2526, w)
    lambda_liv  = recency_weighted_xg(Liv_2425,  n_2425, Liv_2526,  n_2526, w)
    rows.append({
        "w": w,
        "Lambda_City": lambda_city,
        "Lambda_Liv": lambda_liv,
    })

df_lambdas = pd.DataFrame(rows)
df_lambdas["City_Scaled"] = 1.2 * 0.75 * df_lambdas["Lambda_City"]
df_lambdas["Liv_Scaled"] = 0.8 * 0.75 * df_lambdas["Lambda_Liv"]

df_lambdas
```

```python
def row_probs(row):

    p_city_win, p_draw, p_liv_win = match_probs(row["City_Scaled"], row["Liv_Scaled"], K=10)
    return pd.Series({
        "P_Liv_Win": p_liv_win,  
        "P_Draw": p_draw,     
        "P_City_Win": p_city_win, 
    })

probs = df_lambdas.apply(row_probs, axis=1)
df_lambdas = pd.concat([df_lambdas, probs], axis=1)

probs

```
